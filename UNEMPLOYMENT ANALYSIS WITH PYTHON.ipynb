```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data = pd.read_csv('C:\\Users\\Abhishek\\Desktop\\UnemploymentinIndia.csv')
data.columns = data.columns.str.strip()
data[' Date'] = pd.to_datetime(data['Date'], format=' %d-%m-%Y')
plt.figure(figsize=(10,6))
sns.lineplot(x='Date', y='Estimated Unemployment Rate (%)', data=data, errorbar=None)
plt.title('Unemployment Rate in India Over Time')
plt.xlabel('Date')
plt.ylabel('Unemployment Rate (%)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
plt.figure(figsize=(12,8))
sns.boxplot(x='Region', y='Estimated Unemployment Rate (%)', data=data)
plt.title('Regional Comparison of Unemployment Rates in India')
plt.xlabel('Region')
plt.ylabel('Unemployment Rate (%)')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
plt.figure(figsize=(8,6))
sns.boxplot(x='Area', y='Estimated Unemployment Rate (%)', data=data)
plt.title('Rural vs Urban Comparison of Unemployment Rates')
plt.xlabel('Area')
plt.ylabel('Unemployment Rate (%)')
plt.tight_layout()
plt.show()
```
